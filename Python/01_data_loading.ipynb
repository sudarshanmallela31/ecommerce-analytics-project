=====================================

```python
#Dataset Overview
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.columns
```

```python
#Check Data Types
df.dtypes
```

```python
#Missing Values
df.isnull().sum()
```

```python
#Duplicate Records
df.duplicated().sum()
```

```python
#Basic Business KPIs
print("Total Sales :", round(df["Sales"].sum(),2))
print("Total Profit :", round(df["Profit"].sum(),2))
print("Total Orders :", df["Order ID"].nunique())
print("Total Customers :", df["Customer ID"].nunique())
```

```python
#Top Categories
df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

```python
#Top Products
df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(10)
```

```python
#Worst Products
df.groupby("Product Name")["Profit"].sum().sort_values().head(10)
```

```python
#Top Customers
df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
#Regional Analysis
df.groupby("Region")[["Sales","Profit"]].sum().sort_values("Sales", ascending=False)
```

```python
#Profit Margin by Category
(
df.groupby("Category")[["Sales","Profit"]]
.sum()
.assign(Profit_Margin=lambda x: round((x["Profit"]/x["Sales"])*100,2))
)
```

```python
#Discount Impact
df.groupby("Discount")[["Sales","Profit"]].sum()
```

```python
#Monthly Sales Trend
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Month"] = df["Order Date"].dt.month_name()
df.groupby("Month")["Sales"].sum().sort_values(ascending=False)
```

```python
month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly_sales = (
    df.groupby("Month")["Sales"]
    .sum()
    .reindex(month_order)
)

monthly_sales
```

========================================================================
                            **Visualization**
========================================================================

```python
#Sales by Category
import matplotlib.pyplot as plt
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.ylabel("Sales")
plt.xlabel("Category")

plt.show()
```

```python
#Profit by Category
category_profit = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(8,5))
category_profit.plot(kind="bar")

plt.title("Profit by Category")
plt.ylabel("Profit")
plt.xlabel("Category")

plt.show()
```

```python
#Top 10 Customers
top_customers = (
    df.groupby("Customer Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
top_customers.plot(kind="bar")

plt.title("Top 10 Customers by Sales")
plt.ylabel("Sales")

plt.show()
```

```python
#Monthly Sales Trend
month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]
monthly_sales = (
    df.groupby("Month")["Sales"]
      .sum()
      .reindex(month_order)
)
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.grid(True)
plt.show()
```

```python
#Regional Sales Analysis
region_sales = (
    df.groupby("Region")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

region_sales.plot(kind="bar")

plt.title("Sales by Region")
plt.ylabel("Sales")
plt.xlabel("Region")

plt.show()
```

```python
#Regional Profit Analysis
region_profit = (
    df.groupby("Region")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

region_profit.plot(kind="bar")

plt.title("Profit by Region")
plt.ylabel("Profit")
plt.xlabel("Region")

plt.show()
```
