```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(
    r"C:\Users\SudarshanMallea\Desktop\Ecommerce-Analytics-Project\Dataset\Sample - Superstore.csv.csv",
    encoding="cp1252"
)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

monthly_sales = (
    df.groupby(df["Order Date"].dt.month_name())["Sales"]
    .sum()
)

month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly_sales = monthly_sales.reindex(month_order)

plt.figure(figsize=(10,5))
monthly_sales.plot(marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
#Sales by Category
sales_cat = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
sales_cat.plot(kind="bar")
plt.title("Sales by Category")
plt.ylabel("Sales")
plt.show()
```

```python
#Profit by Category
profit_cat = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(8,5))
profit_cat.plot(kind="bar")
plt.title("Profit by Category")
plt.ylabel("Profit")
plt.show()
```

```python
#Sales by Region
sales_region = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))
sales_region.plot(kind="bar")
plt.title("Sales by Region")
plt.ylabel("Sales")
plt.show()
```

```python
#Monthly sales Trend
df["Order Date"] = pd.to_datetime(df["Order Date"])

monthly_sales = (
    df.groupby(df["Order Date"].dt.month)["Sales"]
    .sum()
)

plt.figure(figsize=(10,5))
monthly_sales.plot(marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
